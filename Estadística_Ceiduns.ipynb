**Importando Librerias**



```python

import seaborn as sns
import pandas as pd
df = pd.read_csv('2018_notas_ceiduns_Julio.csv')
```

**Preprocesamiento de Datos - Verificar la existencia de datos perdidos**

```python
# Verificar existencia de datos perdidos
sns.heatmap(df.isnull(), cbar=False)
df.isnull().sum()
```

**Preprocesamiento de Datos - Inputaci√≥n de datos perdidos**

```python
# Inputaci√≥n de valores perdidos
# En caso de variables categ√≥ricas, utilizamos la Moda. Caso contrario, utilizaremos la media

# En este caso, como es manejable el dato perdido de la nota final en la fila del estudiante que no asisti√≥ a clases, la imputamos con 0
df['Final Mark'].fillna(0, inplace=True)
df
```

**Calcular la Media, Mediana y Moda de las notas finales**

```python
# La media es la suma de N elementos dividido entre N, es susceptible a los extremos
# La mediana es el valor dentro del conjunto de datos que es menor que el 50% de los datos y mayor que el 50% restante (Valor Central)
# La moda es el valor mas repetido
media = df["Final Mark"].mean()
mediana = df["Final Mark"].median()
moda = df["Final Mark"].mode()
print("""
    Media: %d
    Mediana: %d
    Moda: %d
""" % (media,mediana,moda))
```

**¬øCu√°l es la nota que tiene como m√≠nimo el 10% m√°s nota de la clase?***

```python
# Los cuartiles abarcan el 25%, 50% y 75% de las observaciones. Los percentiles son una generalizaci√≥n con cualquier porcentaje
p90 = df["Final Mark"].quantile(0.9)
p90

# Los cuartiles dividen el dato en 4 partes iguales (1/4 de los datos o  25%)
# El segundo cuartil coincide con la mediana

# El Decil divide los datos en 10 partes iguales
# El Perceptil dibide los datos en 100 partes iguales

# El Perceptil 90 contiene el 10% de los datos
# El resultado de que el 10% con mas nota de la clase ha sacado como minimo 15.3
```
